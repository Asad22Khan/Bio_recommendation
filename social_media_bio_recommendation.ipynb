```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2") 
```

```python
df["embeddings"] = df["bio"].apply(lambda x: model.encode(x,show_progress_bar=True))
```

```python
def recommend_bio(query, top_k=3):
    # Encode query
    query_vec = model.encode([query])
    
    # Stack embeddings into matrix
    bio_matrix = np.vstack(df["embeddings"].values)
    
    # Compute cosine similarity
    similarities = cosine_similarity(query_vec, bio_matrix)[0]
    
    # Attach similarity scores to df
    df["similarity"] = similarities
    
    # Sort and return top_k recommendations
    results = df.sort_values(by="similarity", ascending=False).head(top_k)
    return results[["id", "bio", "similarity"]]

# -----------------------
# Example Search
# -----------------------
query = "Wanderlust"
recs = recommend_bio(query, top_k=10)

# Iterate properly over DataFrame rows
for _, r in recs.iterrows():
    print(f"User_ID: {r['id']}, Similarity: {r['similarity']:.4f}")
    print(f"Bio: {r['bio']}\n")
```
